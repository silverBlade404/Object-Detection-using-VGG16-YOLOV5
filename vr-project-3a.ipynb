```python
import numpy as np 
import pandas as pd 

import tensorflow as tf 
from tensorflow.keras.datasets import cifar10
import cv2 
```

```python
(x_train , y_train) , (x_test , y_test )= cifar10.load_data()
```

```python
def normalize(matrix): 
    ans = []
    for i in matrix  : 
        i = i/255.0 
        ans.append(i)
    return ans 
x_train_norm = normalize(x_train)
x_test_norm = normalize(x_test)
```

```python
from tensorflow.keras.layers import Input, Conv2D, Dense, Flatten, Dropout
from tensorflow.keras.layers import MaxPooling2D
from tensorflow.keras.layers import BatchNormalization
from tensorflow.keras.models import Sequential
from keras.callbacks import ModelCheckpoint
```

```python
def Custom_Model(activation , conv , dense ) :
    opt = tf.keras.optimizers.SGD(learning_rate=0.1, momentum=0.9)
    if conv == 5 and dense ==3 :
        model = Sequential()
        model.add(Conv2D(input_shape= (32,32,3) , filters = 64, kernel_size =  (5, 5), padding = 'same'))
        model.add(BatchNormalization()) 
        model.add(MaxPooling2D((2, 2)))

        model.add(Conv2D(128, (5, 5),activation=activation  , padding = 'same'))
        model.add(BatchNormalization()) 
        model.add(MaxPooling2D((2, 2)))

        model.add(Conv2D(256, (3, 3),activation=activation , padding = 'same' ))
        model.add(BatchNormalization()) 
        model.add(MaxPooling2D((2, 2))) 

        model.add(Conv2D(512, (3, 3),activation=activation , padding = 'same' ))
        model.add(BatchNormalization()) 
        model.add(MaxPooling2D((2, 2)))

        model.add(Conv2D(512, (3, 3),activation=activation , padding = 'same' ))
        model.add(BatchNormalization()) 
        model.add(MaxPooling2D((2, 2)))

        model.add(Flatten())

        model.add(Dense(4096 ,activation=activation))
        model.add(Dropout(0.2))

        model.add(Dense(4096,activation=activation))   
        model.add(Dropout(0.2))

        model.add(Dense(10 ))
        

        model.compile(optimizer = opt ,loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
                      metrics=['accuracy'])

        return model
    
    elif conv == 4 and dense == 3 : 
        model = Sequential()
        
        model.add(Conv2D(input_shape= (32,32,3) , filters = 64, kernel_size =  (5, 5), padding = 'same'))
        model.add(BatchNormalization()) 
        model.add(MaxPooling2D((2, 2)))

        model.add(Conv2D(128, (5, 5),activation=activation  , padding = 'same'))
        model.add(BatchNormalization()) 
        model.add(MaxPooling2D((2, 2)))

        model.add(Conv2D(256, (3, 3),activation=activation , padding = 'same' ))
        model.add(BatchNormalization()) 
        model.add(MaxPooling2D((2, 2))) 

        model.add(Conv2D(512, (3, 3),activation=activation , padding = 'same' ))
        model.add(BatchNormalization()) 
        model.add(MaxPooling2D((2, 2)))


        model.add(Flatten())

        model.add(Dense(4096 ,activation=activation))
        model.add(Dropout(0.2))

        model.add(Dense(4096,activation=activation))   
        model.add(Dropout(0.2))

        model.add(Dense(10 ))
        

        model.compile(optimizer = opt ,loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
                      metrics=['accuracy'])
        return model 
    
    else : 
        model = Sequential()
        model.add(Conv2D(input_shape= (32,32,3) , filters = 64, kernel_size =  (5, 5), padding = 'same'))
        model.add(BatchNormalization()) 
        model.add(MaxPooling2D((2, 2)))

        model.add(Conv2D(128, (5, 5),activation=activation  , padding = 'same'))
        model.add(BatchNormalization()) 
        model.add(MaxPooling2D((2, 2)))


        model.add(Conv2D(512, (3, 3),activation=activation , padding = 'same' ))
        model.add(BatchNormalization()) 
        model.add(MaxPooling2D((2, 2)))


        model.add(Flatten())

        model.add(Dense(4096 ,activation=activation))
        model.add(Dropout(0.2))

        model.add(Dense(4096,activation=activation))   
        model.add(Dropout(0.2))

        model.add(Dense(10 ))
        

        model.compile(optimizer = opt ,loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
                      metrics=['accuracy'])
        return model 
        
```

```python
# Taking the test set as the validation set for each epoch to calculate accuracy for the test set directly 
print("Training model with 5 CNN and 3 Dense layers : \n")

model_relu_5_3 = Custom_Model('relu' ,5 ,3)
history_relu_5_3 = model_relu_5_3.fit(x_train , y_train ,epochs=10, 
                            validation_data=(x_test, y_test) )

print("\nTraining model with 4 CNN and 3 Dense layers : \n")
model_relu_4_3 = Custom_Model('relu' ,4 ,3)
history_relu_4_3 = model_relu_4_3.fit(x_train , y_train ,epochs=10, 
                            validation_data=(x_test, y_test) )

print("\nTraining model with 3 CNN and 3 Dense layers : \n")
model_relu_3_3 = Custom_Model('relu' ,3 ,3)
history_relu_3_3 = model_relu_3_3.fit(x_train , y_train ,epochs=10, 
                            validation_data=(x_test, y_test) )
```

```python
print("Training model with 5 CNN and 3 Dense layers : \n")
model_tanh_5_3 = Custom_Model('tanh' , 5 , 3)
history_tanh_5_3 = model_tanh_5_3.fit(x_train , y_train ,epochs=10, 
                            validation_data=(x_test, y_test))

print("\nTraining model with 4 CNN and 3 Dense layers : \n")
model_tanh_4_3 = Custom_Model('tanh' , 4 , 3)
history_tanh_4_3= model_tanh_4_3.fit(x_train , y_train ,epochs=10, 
                            validation_data=(x_test, y_test))

print("Training model with 3 CNN and 3 Dense layers : \n")
model_tanh_3_3 = Custom_Model('tanh' , 3 , 3)
history_tanh_3_3 = model_tanh_3_3.fit(x_train , y_train ,epochs=10, 
                            validation_data=(x_test, y_test))
```

```python
print("Training model with 5 CNN and 3 Dense layers : \n")
model_sigmoid_5_3 = Custom_Model('sigmoid' , 5,3 )
history_sigmoid_5_3 = model_sigmoid_5_3.fit(x_train , y_train ,epochs=10, validation_data=(x_test, y_test) )

print("Training model with 4 CNN and 3 Dense layers : \n")
model_sigmoid_4_3 = Custom_Model('sigmoid' , 4,3 )
history_sigmoid_4_3 = model_sigmoid_4_3.fit(x_train , y_train ,epochs=10, validation_data=(x_test, y_test) )

print("Training model with 3 CNN and 3 Dense layers : \n")
model_sigmoid_3_3 = Custom_Model('sigmoid' , 3,3 )
history_sigmoid_3_3= model_sigmoid_3_3.fit(x_train , y_train ,epochs=10, validation_data=(x_test, y_test) )
```

#### Comparing activation function performance on three different models 

```python
import matplotlib.pyplot as plt 

plt.figure(figsize=(7,7))

plt.plot(range(1 , 11) ,history_tanh_5_3.history['val_accuracy'] , 'r-' , range(1 , 11),history_tanh_4_3.history['val_accuracy']  , 'b-',range(1 , 11),history_tanh_3_3.history['val_accuracy'] ,'k-' )
plt.xlabel('epochs')
plt.ylabel('accuracy')
plt.title('TanH activation function')
plt.legend(["Conv = 5 , Dense = 3", "Conv = 4 , Dense = 3" ," Conv = 4 , Dense = 3"], loc ="lower right")
plt.savefig('tanh.png')

plt.figure(figsize=(7,7))

plt.plot(range(1 , 11) ,history_sigmoid_5_3.history['val_accuracy'] , 'r-' , range(1 , 11),history_sigmoid_4_3.history['val_accuracy']  , 'b-',range(1 , 11),history_sigmoid_3_3.history['val_accuracy'] ,'k-' )
plt.xlabel('epochs')
plt.ylabel('accuracy')
plt.title('sigmoid activation function')
plt.legend(["Conv = 5 , Dense = 3", "Conv = 4 , Dense = 3" ," Conv = 4 , Dense = 3"], loc ="lower right")
plt.savefig('sigmoid.png')

plt.figure(figsize=(7,7))

plt.plot(range(1 , 11) ,history_relu_5_3.history['val_accuracy'] , 'r-' , range(1 , 11),history_relu_4_3.history['val_accuracy']  , 'b-',range(1 , 11),history_relu_3_3.history['val_accuracy'] ,'k-' )
plt.xlabel('epochs')
plt.ylabel('accuracy')
plt.title('ReLu activation function')
plt.legend(["Conv = 5 , Dense = 3", "Conv = 4 , Dense = 3" ," Conv = 4 , Dense = 3"], loc ="lower right")
plt.savefig('relu.png')
```
